```python
#Reg Exp to search the email id
import re
email = '''dharani.r.work@gmail.com
           pk.r123@gmail.com
           abc123@gmail.com
           8088656971'''
res = re.search(r"[a-zA-Z0-9.@_]",email)
print(res)
res = re.findall(r"[a-zA-Z0-9\.]+@[a-z]+\.[a-z]+",email)
print(res)
```

```python
#substitute function which will replaces the existing test fun
email = '''dharani.r.work@gmail.com
           pk.r123@skyllx.com
           abc123@insta.com
           8088656971'''
res = re.sub(r"@[a-z]+\.[a-z]+","gmail.com",email)# "@[a-z]+\.[a-z]+" used to change the value and changes the skyllx.com to the gmail.com
print(res)
res = re.sub(r"([a-zA-Z0-9\.]+)(@[a-z]+\.[a-z]+)",r"name:\1,domain:\2", email)#prints name and domain separately
print(res)
```

```python
#few functions from the group objects
s = "skyllx for tech"
res = re.search(r"skyllx",s)
print(res)
print(res.start())
print(res.end())
print(res.group())
```

```python
# split function used to split  the words of where u want to split ,and on what u want to split
s = "sky@123@gmail.com@aaaa@bbbb"
print(re.split(r"@",s))
```

```python
#By using compile fun u can wrap up reg exp into a variable use it multiple timea
s = '''dharani.r.work@gmail.com
           pk.r123@skyllx.com
           abc123@insta.com
           8088656971'''
regex = re.compile(r"[a-zA-Z0-9\.]+@[a-z]+\.[a-z]+")
res = regex.search(s)
print(res)
res = regex.findall(s)
print(res)
```
