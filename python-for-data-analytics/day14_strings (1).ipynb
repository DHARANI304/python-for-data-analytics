```python
#converting strings to binary format
a = 45
b = 16
print("a={:b} b={:b}".format(a,b))

y=213.2415674
print("{:e}".format(y))
pi = 3.1453789623
print("{:.2f}".format(pi))#take 2 digit float number
y =0.25
print("{:.2%}".format(y))# gives percentage of the number
```

```python
#% SPECIFIERS
a=20.25
print("%d"%(a))
b = 13
print("%.2f"%(b))
a = 5
print("%x"%(a))
print("%s"%(a))
print("%o"%(a))
```

```python
a = "Hello World a e i o u"
print(a.lower())
print(a.upper())
print(a.capitalize())
print(a.title())
print(a.swapcase())# it will convert it into small if it is captital or viseversa
print(a.index('l'))#gives index of the value eg:if we pass l
table = a.maketrans("aeiou","12345")
print(table)
print(a.translate(table))
# reverse a string
a = "hello world"
for i in reversed(a):
    print(i)
# using split funtion
a = "welcome to tech"
print(a.split(","))
print(a.split("/"))
b =[]
for i in reversed(a):
    c = b.append(i)
print(c)

# Membership check
a = "welcome to tech"
print(a.startswith('w'))
print(a.endswith('h'))

#compare both the stringd
s1 = "skyllx"
s2 = "tech"
if(s1==s2):
    print("string values are equal")
else:
    print("string values are not equal")
# check id of a strings
s1 = "skyllx"
s2 = "tech"
if(s1 is s2):
    print("ids are same for both values")
else:
    print("ids are not same for both values")

#String concatenation
s1 = "skyllx"
s2 = "tech"
s3 = s1+s2
print(s3)
s3 = ("".join([s1,s2]))
print(s3)
s3 = "{}{}".format(s1,s2)
print(s3)
s3 = f"{s1}{s2}"
```
