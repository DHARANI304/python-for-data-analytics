```python
import time
size = 1000
a = list(range(size))
b = list(range(size))
def dotProduct(a,b):
    ans = 0
    for i,j in zip(a,b):
        ans+= i*j
    return ans
start = time.perf_counter()
for i in range(size):
    dotProduct(a,b)
end = time.perf_counter()
x = np.array(range(size))
y = np.array(range(size))
print(f"time took to do dot product using python list {end-start:.4f}")
start = time.perf_counter()
for i in range(size):
    dotProduct(a,b)
end = time.perf_counter()
print(f"time took to do dot product using numpy array {end-start:.4f}")
```

```python
#SORTING OF AN ARRAY
#* Sorting will happen in ascending order if there are numbers
#* sif there are strings then it will happen based on the alphabets
#*uses quick sort and the ascending order

import numpy as np
a = np.array([80,70,90,60,40,50,30,10,20])
b=np.sort(a)
print(f"the sorted array is:\n{b}")
c = np.sort(-a)
print(f"the sorted array in the negative values is:\n{c}")
d = np.sort(a, axis=0)
print(f"the sorted array with the axis is:\n{d}")
arr = np.array(["Rashmika","anushka","katrina","vijay","zoy"])
np.sort(arr)
# sorting 2d array using axis
arr = np.array([[40,30,20],[70,60,80],[100,90,10]])
d = np.sort(arr, axis=0)#uding axis 0
print(f"the sorted array with the axis 0 is:\n{d}")
arr = np.array([[40,30,20],[70,60,80],[100,90,10]])
d = np.sort(arr, axis=1)#uding axis 1
print(f"the sorted array with the axis 1 is:\n{d}")
```

```python
names = [("names","S10"),("age",int),("marks",float)]
values = [("ram",21,97.5),("shyam",20,87.65),("Azad",22,99.8)]
arr = np.array(values, dtype=names)
print(arr)
c = np.sort(arr)
print(f"the sorted array:\n{c}")
d = np.sort(arr, order = "age")
print(f"the sorted array using age:\n{c}")
```

```python
#WHERE function
#* used for searching elements inside array
#* np.where(condition,[x,y])
#* if cond matches then element replaced by x ,if not then y
#* it will return indexes of elemets
#*if it 2d array then it gives (rows),(columns)
arr = np.array([1,2,3,1,2,3,4,5,6,7,8])
b= np.where(arr==2)
print(f"result which is equal to 2:\n{b}")
print(arr[b])
res =np.where(arr%2==0,"even","odd")
print(res)
res1 =np.where(arr%2==0,2,1)
print(res1)
```

```python
# insert function
#*used to insert elements in an array
arr = np.array([10,20,30,40])
res = np.insert(arr, 1,99)
print(res)
#2d array
arr = np.array([[10,20],[30,40]])
print(arr)
res = np.insert(arr, 1,100, axis = 0)
print(res)
     
```

```python
#append function
a = np.array([10,20,30,40])
b = np.append(a,99)
print(b)
arr = np.array([[10,20],[30,40]])
b = np.append(arr,[[99,100]],axis =0)
print(b)
```

```python
#delete function
import numpy as np
arr = np.array([1,2,3,4,5,6])
res = np.delete(arr, 4)# here 4 is the index no
print(f"after deletion is:\n{res}")

#2D array
#we use axis if we want to delete the portion of elements
a = np.arange(1,10).reshape(3,3)
print(f"the original array :\n{a}")
res = np.delete(a,1,axis = 0)
print(f"after deletion is:\n{res}")

#1d array
#if we want to delete portion of elements we use the range and sliceing
arr = np.array([1,2,3,4,5,6])
res = np.delete(arr, range(1,5))
res1 = np.delete(arr, np.s_[1:5])
print(f"after deletion is:\n{res}")
print(f"after deletion is:\n{res1}")
```

```python
#3D array
a = np.arange(1,61).reshape(3,4,5)
print(f"the original array:\n{a}")
res = np.delete(a,0,axis = 0)
print(f"after deletion is:\n{res}")
```
