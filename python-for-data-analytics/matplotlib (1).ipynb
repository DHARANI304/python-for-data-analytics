```python
# we can have a linestyle,marker,color togethet by using MLC or clm
#MLC
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
plt.plot(x,y,'s--g')
plt.title("line plot")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
#clm
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
plt.plot(x,y,'b-.d')# 'b--d' where b- blue as color, -- linestyle, d-marker
plt.title("line plot")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
#alpha property
# alpha value lies between the 0.0 - 1.0
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
plt.plot(x,y,'s-.k', alpha= 1.0) #alpha = 0.0 wont give a line but 1.0 gives a line
plt.title("line plot")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
# line width shows the width of the line(ls)
# marker size shows the size of the marker (ms)
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
plt.plot(x,y, 'h-g', alpha = 1.0, lw=15, ms=20)
plt.title("line plot")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
# MFC
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
plt.plot(x,y, 'h-.g', alpha = 1.0, lw=15, ms=20, mfc = "red")
plt.title("line plot")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
# ploting of the figure

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,21)
y = x**2
y1 = x**3
plt.figure(figsize =(12,6),facecolor="lightgreen")
           
plt.plot(x,y, 'h-.g', alpha = 1.0, mfc = "g")
plt.plot(x,y1, 'o-r', alpha = 1.0,  mfc = "r")
plt.title("line plot with figure size and face colour")
plt.xlabel("values")
plt.ylabel("square of x")
```

```python
# plotting one graph with multiple linew
x = np.arange(1,11)
i = x;
s = x**2;
c = x**3;
plt.plot(x,i,"o-r", x,s,"s-.y", x,c,"o:b")
plt.title("one graph for many plots")
plt.show()
```

```python
# plotting one graph with multiple lines
#in title we use location,padding,vertical position of title
x = np.arange(1,11)
i = x;
s = x**2;
c = x**3;
plt.plot(x,i,"o-r", x,s,"s-.y", x,c,"o:b")
plt.title("one graph for many plots", loc = "left", pad = 20, y = 1.05, color="g")
plt.show()
```
