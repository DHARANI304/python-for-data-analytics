```python
import re
s = '''91-8088656971
       45-7866907689'''
res = re.sub(r"[0-9]-","91",s)
print(res)
```

```python
import re
s = "skyllx for tech"
res = re.findall(r"\b\w{2}",s)
print(res)
```

```python
# write a re to match the first word of a string
s = "skyllx for tech"
res = re.match(r"skyllx",s)
print(res)
```

```python
s = '''9808111112
       krishakat
       02/10/1994
       kk@gmail.com'''
res = re.search(r"[0-9]+\/[0-9]+\/[0-9]+", s)
print(res)
```

```python
# write a re to find the indian pincode first number should start [1-9]
s ='''5678904
      560098
      560876
      4567890'''
res = re.findall(r"\b[1-9][0-9]{5}\b",s)
print(res)
```

```python
# FLAGS in regular expressions,
s = "skyllx for tech"
res = re.search(r"[F][O][R]",s, re.IGNORECASE)# Ignorecase will ignore the case of the words
print(res)
res = re.search(r"[F][O][R]",s, re.I)# even single I  will ignore the case of the words
print(res)
```

```python
#DOTALL Function will give all the words
s = "skyllx.\n for.\n tech"
res = re.search(r".+",s)# it will match only skyllx.
print(res)
res1 = re.search(r".+",s,re.DOTALL)
print(res1)
```

```python
#MULTILINE Flag used to match the values
s = "skyllx.\n for.\n tech"
res = (r"^tech",s,re.MULTILINE)
print(res)
```

```python
s = '''<html></html>
       <head></head>
       <title></title>'''
res = re.findall(r"<\/[a-zA-Z][a-zA-Z0-9]*>", s)
print(res)
res1 = re.findall(r"<[a-zA-Z][a-zA-Z]*>",s)
print(res1)
```
