```python
# Working on the dictionaries using zip function
#Zip combines two or more iterable elements into the tuples
names = ['Dharani','sahana','anjali']
age = [22,20,21]
student = dict(zip(names,age,))
print(student)
for i in zip(names,age,):
    print(i)
```

```python
# all function :It will check if all the elements are ture then it will give true .if any one element is false it will return false
lst = [True,True,True]
res = all(lst)
print(res)

lst2 = [0==0,1==1,2==3]
res = all(lst2)
print(res)
```

```python
#any function checks if any one element is true
lst = [True,True,True]
res = any(lst)
print(res)

lst2 = [0==0,1==1,2==3]
res = any(lst2)
print(res)
```
