```python
# Adding/updating the values in the dictionary there are 4 types where we can update a dictionaries
a = {1:"a",2:"b",3:"c",}
print(a)
a[4]="d"
a.update({5:"e",6:"f",7:"g"})
print(a)
a.update([{8:"h",9:"i"}])
print(a)
a.update(one="a")
print(a)
```

```python
#Accessing the elements of the dictionaries
d = {1:"a",2:"b",3:[10,20,30],4:"d"}
print(d)
print(type(d))
print(d[2])
lst = d[3]
print(lst)
lst.append(99)
print(lst)
print(d)
print(d.get(4)) #to access values we use  get function
#Deleting the values
d[1]="dharani" #updating the  value
print(d)
print(d.pop(2))
d.popitem
print(d)
del d[1]
d.clear()
print(d)
```

```python
#Accessing keys,values,items
d = {1:"a",2:"b",3:"c"}
print(d)
print(type(d))
print(d.keys())
print(d.values())
print(d.items())
for i in d.keys():
    print(i)
for i in d.values():
    print(i)
```

```python
#Membership functions
D = {1:"a", 2:"b",3:"c"}
print(1 in D)
print('a' in D.values())
print((1,'a')in D.items())
```

```python
# Concatination of the dictionary
d1 = {1:"a",2:"b",3:"c"}
d2 = {4:"d",5:"e"}
d3 = {**d1,**d2}
print(d3)
```

```python
# power of the values
a = [1,2,3,4,5]
d = dict()
for i in a:
    d[i] = i**2
print(d)
# Printing only even numbers
d2 = {i:i**2 for i in a if i%2==0}
print(d2)
#If the element is odd then print cube else print square
for i in a:
    if i%2==0:
        d[i]=i**2
    else:
        d[i]=i**3
print(d)

d3 ={i:i**2 if i%2==0 else i**3 for i in a}
print(d3)
```
